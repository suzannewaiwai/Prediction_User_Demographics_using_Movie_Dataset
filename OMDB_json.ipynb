```python
def clean_the_title(title):
    # Check if the title is a string and ends with ", The"
    if isinstance(title, str) and title.endswith(", The"):
        return title.replace(", The", "")
    return title
```

```python
import pandas as pd
import requests

movies_path = 'movies.dat'
# Load your MovieLens dataset (e.g., a CSV file)
# Replace with the path to your MovieLens file
movies_df = pd.read_csv(movies_path, sep='::', header=None, names=['MovieID', 'Title', 'Genres'], encoding='ISO-8859-1')

movies_df['Year'] = movies_df['Title'].str.extract(r'\((\d{4})\)', expand=False)
movies_df['Clean_Title'] = movies_df['Title'].str.extract(r'^(.*) \(\d{4}\)$', expand=False)
movies_df['Cleaned_Title'] = movies_df['Clean_Title'].apply(clean_the_title)
    
movies_df.head()
```

```python
# Create a list to store movie data
movie_data_list = []

# Loop through each movie in the dataset
for index, row in movies_four.iterrows():
    movieid = row['MovieID']
    title = row['Cleaned_Title']
    year = row['Year']  # Adjust the column name according to your dataset

    # Fetch data from OMDb API
    movie_data = fetch_movie_data(title, year)
    
    if movie_data and movie_data.get('Response') == 'True':
        # Extract relevant information
        movie_info = {
            'MovieID' : movieid,
            'Title': movie_data.get('Title'),
            'Year': movie_data.get('Year'),
            'Rated': movie_data.get('Rated'),
            'Released': movie_data.get('Released'),
            'Runtime': movie_data.get('Runtime'),
            'Genre': movie_data.get('Genre'),
            'Director': movie_data.get('Director'),
            'Writer': movie_data.get('Writer'),
            'Actors': movie_data.get('Actors'),
            'Plot': movie_data.get('Plot'),
            'Language': movie_data.get('Language'),
            'Country': movie_data.get('Country'),
            'Awards': movie_data.get('Awards'),
            'Poster': movie_data.get('Poster'),
            'Metascore': movie_data.get('Metascore'),
            'imdbRating': movie_data.get('imdbRating'),
            'imdbVotes': movie_data.get('imdbVotes'),
            'imdbID': movie_data.get('imdbID'),
            'Type': movie_data.get('Type'),
            'DVD': movie_data.get('DVD'),
            'BoxOffice': movie_data.get('BoxOffice'),
            'Production': movie_data.get('Production'),
            'Website': movie_data.get('Website')
        }
        movie_data_list.append(movie_info)

# Create a DataFrame from the list of movie data
movies_omdb_df = pd.DataFrame(movie_data_list)


# Save the data to a CSV file (optional)
movies_omdb_df.to_csv('movies_four.csv', index=False)

# Display the first few rows of the DataFrame
print(movies_four.head())
```
